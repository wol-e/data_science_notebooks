```python
def show_meals(short=True):
    if short:
        print([x for x in meals.keys()])
    else:
        print(meals)

show_meals(True)
```

```python
def get_nutrition_essential(meal, printmode=False):
    scaling_factor = meal[1]/meals[meal[0]]["amount"]
    
    if printmode:
        nutrition_string = ""
        for nutrition in meals[meal[0]]['nutrition']:
            nutrition_string += \
                f"\n\t{nutrition['name']}: {round(nutrition['amount'] * scaling_factor,2)} {NUTRITION_ELEMENTS[nutrition['name']]['unit']}\n"
        print(
            f"Nutrition for meal '{meal[0]}' per {meal[1]} {meals[meal[0]]['unit']}:",
            nutrition_string,
            sep="\n"
        )
    
    nutrition_list = copy.deepcopy(meals[meal[0]]['nutrition'])
    for nutr in nutrition_list:
        nutr['amount'] = round(nutr['amount'] * scaling_factor, 2)
    return(nutrition_list)
```

```python
def get_essentials_of_meal(meal):
    """
    meal: tuple of type (string, float) of meal string and amount in g
    
    returns list of essential meals
    
    >>> get_essentials(("potato with orange juice", 200))
    >>> [("potato", 180), ("orange juice", 20)]
    """
    
    def split_essentials_step(ingredients, essentials):
        """
        >>>
        """
        if meals[ingredients[0][0]]["essential"]:
            essentials.append(ingredients[0])

        else:
            recipe = []
            for item in meals[ingredients[0][0]]["recipe"]:
                recipe.append((item[0], item[1] * ingredients[0][1]))
            ingredients = ingredients + recipe

        del ingredients[0]

        return ingredients, essentials

    essentials = []
    ingredients = [meal]
    safety_count = 0
    
    while len(ingredients) > 0:
        ingredients, essentials = split_essentials_step(ingredients, essentials)
        safety_count += 1
        
        if safety_count == 200:
            print("Error, breaking up into essentials did not terminte")
            break

    return essentials
    
```

```python
def get_essentials_of_menu(menu=[], days=1, age="12m"):
    def group_essentials(essentials):
        grouped_essentials = []
        essential_ingredients = set([x[0] for x in essentials])
        
        for x in essential_ingredients:
            amount = 0
            for z in essentials:
                if z[0] == x:
                    amount += z[1]
            grouped_essentials.append((x, amount))
        
        return(grouped_essentials)
    
    essentials = []
    for meal in menu:
        essentials = essentials + (get_essentials_of_meal(meal))

    essentials = group_essentials(essentials)
    
    return essentials
```

```python
def get_nutrition_of_menu(menu=[], printmode=False):
    essentials = get_essentials_of_menu(menu)
    if printmode:
        print(
            f"Menu with units:\n{menu}\n",
            f"Essential ingredients of menu with units:\n{essentials}\n",
            sep="\n"
        )
    essentials_nutrition = []
    for essential in essentials:
        essentials_nutrition = essentials_nutrition + \
            get_nutrition_essential(essential, printmode=printmode)
    
    def group_nutrition(nutrition, printmode=False):
        grouped_nutrition = []
        nutrition_names = set([x['name'] for x in nutrition])
        
        for x in nutrition_names:
            amount = 0
            for z in nutrition:
                if z['name'] == x:
                    amount += z['amount']
            grouped_nutrition.append({'name': x, 'amount': round(amount, 2)})

        if printmode:
            print(f"Total nutrition of menu\n\t{grouped_nutrition}")
        return(grouped_nutrition)
    
    return(group_nutrition(essentials_nutrition))
```

```python
def analyse_nutrition_of_menu(menu_list=[], names= [], age="", logscale=False, percentage=True, print_details=False):
    
    font = {'family': 'DejaVu Sans',
            'weight': 'bold',
            'size': 15}
    
    matplotlib.rc('font', **font)
    
    if age == "":
        print("no age for comparing nutrition requirements provided")
        daily_needs = None
        
    else:
        daily_needs = nutrition_requirement
    
    menu_nbr = len(menu_list)
    
    fig, ax = plt.subplots(menu_nbr, 1, figsize=(18,5 * (menu_nbr + 2)))
    
    for i, menu in enumerate(menu_list):
        
        if menu_nbr == 1:
            axis = ax
        else:
            axis = ax[i]
        
        if names == []:
            menu_name = i
        else:
            menu_name = names[i]
        
        print(f"\n\n### Nutrition info for menu '{menu_name}' ###\n")
        nutrition = get_nutrition_of_menu(menu, printmode = print_details)
        
        error_lower = np.array([nutrition_requirement[age][n["name"]]["lower"] for n in nutrition])
        error_upper = np.array([nutrition_requirement[age][n["name"]]["upper"] for n in nutrition])
        error_mean = (error_upper + error_lower) / 2
        x = [n["name"] for n in nutrition]
        x[x.index("Vitamin B12")] = "Vitamin B12*"
        x[x.index("Iron")] = "Iron**"
                
        y = [n["amount"] for n in nutrition]
        
        print(f" >>> Total Nutrition for menu '{menu_name}':<<<\n",
              f"\tNutrients: {x}",
              f"\tIntake: {y}",
              f"\tMean requirement: {error_mean}\n",
             sep="\n")

        if percentage:
            error_lower = ((error_lower / error_mean) * 100).round(2)
            error_upper = ((error_upper / error_mean) * 100).round(2)
            y = ((y / error_mean) * 100).round(2)
            error_mean = ((error_mean / error_mean) * 100).round(2)

        axis.bar(
            x,
            y,
            label="Nutrition intake", zorder=2
        )

        axis.errorbar(
            x,
            error_mean,
            yerr=[error_mean - error_lower, error_upper - error_mean], fmt='o', color="orange",
            linewidth=6, markersize=12,
            label="Nutrition Recommendation", zorder=3
        )

        axis.legend()
        axis.grid(linestyle='--', zorder=1)
        axis.set_facecolor('whitesmoke')
        
        # axis.set_xticks(fontsize=25)
        # plt.ylabel('ylabel', fontsize=16)
        axis.set_title(f"Nutrition intake for menu '{menu_name}'")
        if percentage:
            axis.set_ylabel("% of mean nutrition recommendation")
        if logscale:
            axis.set_yscale('log')
            axis.set_ylabel('log of amount')
    
    plt.show()
    print(
        """
        * Only certain amounts (<5mcg) of Vitamin B12 can be absorbed per meal without supplements. 
        So the intake of Vitamin B12 may not reflect the amount usable for the body on large portions. See for details: \t
        'https://gesundheit-heute.ch/wp-content/uploads/2017/04/vitamin_b12_faktenblatt-2.pdf'
        \n
        ** There are 2 forms of Iron. The first form from animal meat can be absorbed by the body at a 
        rate of 14 to 18%, while the second from plants can only be absorbed at 5 to 12%.
        So for a plant based diet a higher (maybe 2x) amount of Iron than nutrition 
        recommendtions should be consumed.
        """
         )
```
