# Clustering

```python
from numpy.linalg import norm
from itertools import combinations
pd.options.display.max_rows=2000

parties = list(df_wide.index.values)
distances = []

for pair in combinations(parties, 2):
    distances.append([
        pair[0], pair[1], norm(df_wide.loc[pair[0]] - df_wide.loc[pair[1]]), 0
    ])
    distances.append([
        pair[1], pair[0], norm(df_wide.loc[pair[0]] - df_wide.loc[pair[1]]), 1
    ])
    
distances = pd.DataFrame(distances, columns=["name1", "name2", "distance", "duplicate"])

distances = distances.sort_values(by=["name1", "distance"]).reset_index(drop=True)
distances
```

```python
party="LIEBE"

distances.query(f"name1 == '{party}'")
```

```python
import matplotlib.pyplot as plt

plt.hist(distances.query("duplicate == 0").distance, bins=50)
plt.show()
```

```python
from sklearn.cluster import DBSCAN

cluster = DBSCAN(eps=4, min_samples=2)
cluster.fit(df_wide)

df_wide["dbscan_label"] = cluster.labels_
df_wide.sort_values(by="dbscan_label")
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(
    n_clusters=4,
    affinity="euclidean",
    linkage="ward",
)
    
cluster.fit(df_wide)

df_wide["agglomerative_label"] = cluster.labels_
df_wide.iloc[:,-2:].sort_values(by="agglomerative_label")
```
