```python
import pandas as pd

def read_from_json(subject):
    df = pd.read_json(
        f"qual-o-mat-data/data/2021/deutschland/{subject}.json"
    )
    df = df.set_index("id", drop=True)

    return df


opinion = read_from_json("opinion")
answer = read_from_json("answer")
party = read_from_json("party")
comment = read_from_json("comment")
statement = read_from_json("statement")
```

```python
party.sort_values(by="name")
```

```python
df = opinion\
    .join(party[["name"]], how="left", on="party")\
    .join(statement[["label", "text"]], how="left", on="statement")\
    .join(answer, how="left", on="answer")

df
```

```python
def map_answer(answer):
    answer_map = {0: 1, 1: -1, 2: 0}
    
    return answer_map[answer]
```

```python
df["answer_centered"] = df["answer"].apply(map_answer)
```

```python
"""
Formatting by party and preferences
"""

df_wide = df.copy()[["name", "statement", "answer_centered"]]

df_wide = df_wide.set_index(["name", "statement"]).unstack(level="statement")

df_wide
```
