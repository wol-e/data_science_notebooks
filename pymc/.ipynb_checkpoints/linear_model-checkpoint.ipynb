```python
import pymc3 as pm
import numpy as np
import matplotlib.pyplot as plt
```

```python
size = 200
true_intercept = 1
true_slope = 2

x = np.linspace(0, 1, size)
# y = a + b*x
true_regression_line = true_intercept + true_slope * x
# add noise
y = true_regression_line + np.random.normal(scale=0.5, size=size)

data = dict(x=x, y=y)
```

```python
def plot_data(ax):
    ax.plot(x, y, "x", label="sampled data")
    ax.plot(x, true_regression_line, label="true regression line", lw=2.0)

fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111, xlabel="x", ylabel="y", title="Generated data and underlying model")
plot_data(ax)
plt.legend(loc=0);
```

```python
with pm.Model() as model:  # model specifications in PyMC3 are wrapped in a with-statement
    # Define priors
    sigma = pm.HalfCauchy("sigma", beta=10, testval=1.0)
    intercept = pm.Normal("intercept", 0, sigma=20)
    w = pm.Normal("w", 0, sigma=20)

    # Define likelihood
    likelihood = pm.Normal("y", mu=intercept + w * x, sigma=sigma, observed=y)

    # Inference!
    trace = pm.sample(3000, cores=2)  # draw 3000 posterior samples using NUTS sampling
```

```python
from arviz import plot_trace

plt.figure(figsize=(7, 7))
plot_trace(trace)
plt.tight_layout()
plt.show()
```

```python
from random import choice

fig, ax = plt.subplots(1, 1,figsize=(8, 5))
for i in range(100):
    j = choice(range(len(trace["w"])))
    ax.plot(x, trace["intercept"][j] + trace["w"][j] * x, color='gray', alpha=.2)
ax.set_facecolor("whitesmoke")
ax.grid(True)
plot_data(ax)
```
