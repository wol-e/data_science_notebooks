## Naive Bayes

One problem to vercome when applying Naive Bayes to the data set is the mixture of datatypes (continous and categorical)

```python
from sklearn.naive_bayes import GaussianNB, CategoricalNB
```

```python
df_train_cats = df_train.select_dtypes(include=['object']).copy()
df_train_cats["churn"] = df_train["churn"].values

df_train_cats = df_train_cats.fillna('NaN')

for col in df_train_cats.columns:
    df_train_cats[col] = pd.Categorical(df_train_cats[col])  # TODO: is this necessary?
    df_train_cats[col] = df_train_cats[col].cat.codes

df_train_cont = df_train.select_dtypes(exclude=['object'])

df_train_cats
```

```python
for col in df_train_cont.columns:
    df_train_cont[col] = df_train_cont[col].fillna(df_train_cont[col].mean())

df_train_cont
```

```python
"""Gaussian NB on continous features
- TODO: try standard normlizing features before
"""

model_nb_cont = GaussianNB()

split_index = 40000

y_train_cont = df_train_cont.churn[:split_index]
y_test = df_train_cont.churn[split_index:]

X_train_cont = df_train_cont.iloc[:split_index,:-1]
X_test_cont = df_train_cont.iloc[split_index:,:-1]

model_nb_cont.fit(
    X=X_train_cont,
    y=y_train_cont
)

print(f"""Score on train: {roc_auc_score(
    y_train_cont,
    model_nb_cont.predict_proba(X_train_cont)[:,1],
)} """)

print(f"""Score on test: {roc_auc_score(
    y_test,
    model_nb_cont.predict_proba(X_test_cont)[:,1],
)} """)
```

```python
"""Categorical NB on categorical features
"""

model_nb_cats = CategoricalNB(
    alpha=0.5
)

y_train_cats = df_train_cats.churn[:split_index]

X_train_cats = df_train_cats.iloc[:split_index,:-1]
X_test_cats = df_train_cats.iloc[split_index:,:-1]

# remove unseen categories from test set
def replace_unseen_by_0(x, max_cat_code):
    return x if x <= max_cat_code else 0
    
for col in X_test_cats:
    max_cat_code = X_train_cats[col].max()
    X_test_cats[col] = X_test_cats[col].apply(lambda x: replace_unseen_by_0(x, max_cat_code))

model_nb_cats.fit(
    X=X_train_cats,
    y=y_train_cats
)

print(f"""Score on train: {roc_auc_score(
    y_train_cats,
    model_nb_cats.predict_proba(X_train_cats)[:,1],
)} """)

print(f"""Score on test: {roc_auc_score(
    y_test,
    model_nb_cats.predict_proba(X_test_cats)[:,1],
)} """)
```

```python
"""Combining the models"""

print(f"""Score on test: {roc_auc_score(
    y_test,
    model_nb_cats.predict_proba(X_test_cats)[:,1] * model_nb_cont.predict_proba(X_test_cont)[:,1],
)} """)
```

```python
# Confirming cats model with cross validation
from sklearn.model_selection import cross_validate

y_train_cats = df_train_cats.churn
X_train_cats = df_train_cats.copy().iloc[:,:-1]

model_nb_cats = CategoricalNB(
    alpha=0.5,
    min_categories=(X_train_cats.max() + 1).values
)

cross_validate(
    estimator=model_nb_cats,
    X = X_train_cats,
    y = y_train_cats,
    cv = 5,
    scoring="roc_auc",
    return_train_score=True
)
```

# Train at once on full dataset 

We use one-hot encoding for the categories to be able o train the classifier on the full dataset all being floats. We need to take some care to not simply one-hot encode everything as this would result in thousands of columns. So we will do some thresholding based apporach.


```python
df_train_oh_prep = pd.read_pickle("./data/processed/df_train_for_one_hot.pkl")

df_train_oh_prep
```

```python
y_test = df_train_oh_prep["churn"]
```

```python
original_fields = [col for col in df_train_oh_prep if ("_freq" not in col and "_rank" not in col)]


for col in df_train_oh_prep[original_fields].columns:
    if df_train_oh_prep[col].dtype == "float":
        df_train_oh_prep[col] = df_train_oh_prep[col].fillna(df_train_oh_prep[col].mean())
        
df_train_oh_prep
```

```python
"""
For one hot encoding let's try to to encode all categories which occur more often than a fixed threshold and 
group the rest into a single category
"""

# first filter relevant fields 
df_train_oh = df_train_oh_prep[
    [col for col in df_train_oh_prep.columns if (
        (col in original_fields and df_train_oh_prep[col].dtype == "float")
        or ("_freq" in col)
    )]
]

df_train_oh
```

```python
freq_threshold = 50

for col in df_train_oh.columns:
    if "_freq" in col:
        df_train_oh[col] = df_train_oh[col].apply(lambda x: max(x, freq_threshold))
        
df_train_oh.nunique().sort_values()
```

```python
df_train_oh
```

```python
pd.options.display.max_columns = 1000

df_train_oh = pd.get_dummies(
    df_train_oh,
    columns=[col for col in df_train_oh if "freq" in col]
)

df_train_oh
```

```python
model_nb_oh = GaussianNB()

split_index = 40000

y_train_oh = y_test[:split_index]
y_test_oh = y_test[split_index:]

X_train_oh = df_train_oh.iloc[:split_index,:]
X_test_oh = df_train_oh.iloc[split_index:,:]

model_nb_oh.fit(
    X=X_train_oh,
    y=y_train_oh
)

print(f"""Score on train: {roc_auc_score(
    y_train_oh,
    model_nb_oh.predict_proba(X_train_oh)[:,1],
)} """)

print(f"""Score on test: {roc_auc_score(
    y_test_oh,
    model_nb_oh.predict_proba(X_test_oh)[:,1],
)} """)
```

```python
"""It's not improving, as it would be expected, as the earlier method of combining two models for the 
numeric and categorical variables in theory is the same thing. It's even a little worse, which may be due
to the encoding missing information on the less frequent categories."""
```

```python
"""Let's also save the encoded dataframe to maybe use in other models"""

df = df_train_oh
df["churn"] = y_test

df.to_pickle(f"./data/processed/df_train_oh_freq_th_{freq_threshold}.pkl")
```
