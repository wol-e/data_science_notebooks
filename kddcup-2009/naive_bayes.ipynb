## Naive Bayes

One problem to vercome when applying Naive Bayes to the data set is the mixture of datatypes (continous and categorical)

```python
from sklearn.naive_bayes import GaussianNB, CategoricalNB
```

```python
df_train_cats = df_train.select_dtypes(include=['object']).copy()
df_train_cats["churn"] = df_train["churn"].values

df_train_cats = df_train_cats.fillna('NaN')

for col in df_train_cats.columns:
    df_train_cats[col] = pd.Categorical(df_train_cats[col])  # TODO: is this necessary?
    df_train_cats[col] = df_train_cats[col].cat.codes

df_train_cont = df_train.select_dtypes(exclude=['object'])

df_train_cats
```

```python
for col in df_train_cont.columns:
    df_train_cont[col] = df_train_cont[col].fillna(df_train_cont[col].mean())

df_train_cont
```

```python
"""Gaussian NB on continous features
- TODO: try standard normlizing features before
"""

model_nb_cont = GaussianNB()

split_index = 40000

y_train_cont = df_train_cont.churn[:split_index]
y_test = df_train_cont.churn[split_index:]

X_train_cont = df_train_cont.iloc[:split_index,:-1]
X_test_cont = df_train_cont.iloc[split_index:,:-1]

model_nb_cont.fit(
    X=X_train_cont,
    y=y_train_cont
)

print(f"""Score on train: {roc_auc_score(
    y_train_cont,
    model_nb_cont.predict_proba(X_train_cont)[:,1],
)} """)

print(f"""Score on test: {roc_auc_score(
    y_test,
    model_nb_cont.predict_proba(X_test_cont)[:,1],
)} """)
```

```python
"""Categorical NB on categorical features
"""

model_nb_cats = CategoricalNB(
    alpha=0.5
)

y_train_cats = df_train_cats.churn[:split_index]

X_train_cats = df_train_cats.iloc[:split_index,:-1]
X_test_cats = df_train_cats.iloc[split_index:,:-1]

# remove unseen categories from test set
def replace_unseen_by_0(x, max_cat_code):
    return x if x <= max_cat_code else 0
    
for col in X_test_cats:
    max_cat_code = X_train_cats[col].max()
    X_test_cats[col] = X_test_cats[col].apply(lambda x: replace_unseen_by_0(x, max_cat_code))

model_nb_cats.fit(
    X=X_train_cats,
    y=y_train_cats
)

print(f"""Score on train: {roc_auc_score(
    y_train_cats,
    model_nb_cats.predict_proba(X_train_cats)[:,1],
)} """)

print(f"""Score on test: {roc_auc_score(
    y_test,
    model_nb_cats.predict_proba(X_test_cats)[:,1],
)} """)
```

```python
"""Combining the models"""

print(f"""Score on test: {roc_auc_score(
    y_test,
    model_nb_cats.predict_proba(X_test_cats)[:,1] * model_nb_cont.predict_proba(X_test_cont)[:,1],
)} """)
```

```python
# Confirming cats model with cross validation
from sklearn.model_selection import cross_validate

y_train_cats = df_train_cats.churn
X_train_cats = df_train_cats.copy().iloc[:,:-1]

model_nb_cats = CategoricalNB(
    alpha=0.5,
    min_categories=(X_train_cats.max() + 1).values
)

cross_validate(
    estimator=model_nb_cats,
    X = X_train_cats,
    y = y_train_cats,
    cv = 5,
    scoring="roc_auc",
    return_train_score=True
)
```
