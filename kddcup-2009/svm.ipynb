```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics import roc_auc_score, plot_roc_curve

pd.options.display.max_columns = 200
pd.options.display.max_rows = 100
```

```python
df_train = pd.read_pickle(f"./data/processed/df_train_oh_freq_th_50.pkl")

df_train
```

```python
from sklearn.svm import SVC, LinearSVC
```

```python
help(LinearSVC)
```

```python
y = df_train["churn"]

x_train = df_train.iloc[:, :-1]
```

```python
model = LinearSVC(
    penalty='l2',
    loss='squared_hinge',
    max_iter=5000,
    dual=False,
    C=1,
    class_weight='balanced'
)
```

```python
%%time

cutoff = 10000

model.fit(x_train.iloc[:cutoff,:], y[:cutoff])
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
y_pred = model.predict(x_train.iloc[cutoff:,:])
y_true = y[cutoff:]

print(accuracy_score(y_true, y_pred))

"""
By definition a confusion matrix :math:`C` is such that :math:`C_{i, j}`
    is equal to the number of observations known to be in group :math:`i` and
    predicted to be in group :math:`j`.
    
    Thus in binary classification, the count of true negatives is
    :math:`C_{0,0}`, false negatives is :math:`C_{1,0}`, true positives is
    :math:`C_{1,1}` and false positives is :math:`C_{0,1}`.
"""

confusion_matrix(y_true, y_pred)
```

```python
"""try again with scaling features"""

from sklearn.base import clone
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
```

```python
%%time

model = clone(model)

clf = make_pipeline(StandardScaler(), model)
clf.fit(x_train.iloc[:cutoff,:], y[:cutoff])
```

```python
y_pred = model.predict(x_train.iloc[cutoff:,:])
y_true = y[cutoff:]

print(accuracy_score(y_true, y_pred))

confusion_matrix(y_true, y_pred)
```
